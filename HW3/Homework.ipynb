```python
#task 1
file = open('Spotify Charts.html', 'r', encoding = 'utf-8')
text = file.read()

print(text)
```

```python
#task 2
import re
import json
#the regex is taken from geekforgeeks, i did not write it individually
#fortunatelly, the task does not prohibit in using the internet for finding the
#right regex 
regex = r"(?i)\b((?:https?://|www\d{0,3}[.]|[a-z0-9.\-]+[.][a-z]{2,4}/)(?:[^\s()<>]+|\(([^\s()<>]+|(\([^\s()<>]+\)))*\))+(?:\(([^\s()<>]+|(\([^\s()<>]+\)))*\)|[^\s`!()\[\]{};:'\".,<>?]))"
urls_bad = re.findall(regex, text)
urls = [url[0] for url in urls_bad]

for url in urls:
    print(url)
    
with open('urls.json', 'w') as jsonfile:
    json.dump(urls, jsonfile)
```

```python
#task 3

#At the begining i looked for all colors and there was 7 that have the length 3
#surfing the html file for this 7 length-3 colors, i identified that was not colors
#but the code for -> " -> (&#039;). I changed the regex a little bit, and now
#it selects only the hex colors
regex= r'"#[0-9A-Fa-f]{6}"'
colors = re.findall(regex, text)
colors = [color[1:-1] for color in colors]

for color in colors:
    print(color)

with open('hex.json', 'w') as jsonfile:
    json.dump(colors, jsonfile)
```

```python
#task 4

#After surfing the .html file, i saw that all the names of the artists have
# 'by' in front of the name, so i used this thing in writing the regex.
# For some tracks there are 2 or more interpretters, divided by comma, so i 
#splitted them. Also, some artists have non english letters in their name.

regex = r'>by[a-zA-Z0-9$,. ]+<'
names = re.findall(regex, text)

#removing "by "
names = [name[4:-1] for name in names]

#splitting the artists if there are more authors on the same track
nr = 0
names_new = []
for index in range(len(names)):
    for name in names[index].split(", "):
        names_new.append(name)
names = names_new[:]


for name in names:
    print(name)

with open('artist.json', 'w') as jsonfile:
    json.dump(names, jsonfile)
```
